*Handling Nulls*

```python
def handle_null_values(df):
    # Handle null values for 'Denominations'
    encashed_not_null = df['Denominations(Encashed)'].notnull()
    purchased_null = df['Denominations(Purchased)'].isnull()
    encashed_to_purchased = encashed_not_null & purchased_null
    df.loc[encashed_to_purchased, 'Denominations(Purchased)'] = df.loc[encashed_to_purchased, 'Denominations(Encashed)']

    purchased_not_null = df['Denominations(Purchased)'].notnull()
    encashed_null = df['Denominations(Encashed)'].isnull()
    purchased_to_encashed = purchased_not_null & encashed_null
    df.loc[purchased_to_encashed, 'Denominations(Encashed)'] = df.loc[purchased_to_encashed, 'Denominations(Purchased)']

    # Handle null values for 'Purchase Date'
    encashed_not_null = df['Encashment Date'].notnull()
    purchased_null = df['Purchase Date'].isnull()
    encashed_to_purchased = encashed_not_null & purchased_null
    df.loc[encashed_to_purchased, 'Purchase Date'] = df.loc[encashed_to_purchased, 'Encashment Date']
    
    # Handle null values for 'Purchase Year'
    encashed_not_null = df['Encashment Year'].notnull()
    purchased_null = df['Purchase Year'].isnull()
    encashed_to_purchased = encashed_not_null & purchased_null
    df.loc[encashed_to_purchased, 'Purchase Year'] = df.loc[encashed_to_purchased, 'Encashment Year']
    
    # Handle null values for 'Encashment Year'
    encashed_not_null = df['Purchase Year'].notnull()
    purchased_null = df['Encashment Year'].isnull()
    encashed_to_purchased = encashed_not_null & purchased_null
    df.loc[encashed_to_purchased, 'Encashment Year'] = df.loc[encashed_to_purchased, 'Purchase Year']

    # Handle null values for 'Encashment Date'
    purchased_not_null = df['Purchase Date'].notnull()
    encashed_null = df['Encashment Date'].isnull()
    purchased_to_encashed = purchased_not_null & encashed_null
    df.loc[purchased_to_encashed, 'Encashment Date'] = df.loc[purchased_to_encashed, 'Purchase Date']

    # Handle null values for 'Branch Name'
    encashed_not_null = df['Branch Name(encashed from)'].notnull()
    purchased_null = df['Branch Name(purchased from)'].isnull()
    encashed_to_purchased = encashed_not_null & purchased_null
    df.loc[encashed_to_purchased, 'Branch Name(purchased from)'] = df.loc[encashed_to_purchased, 'Branch Name(encashed from)']

    purchased_not_null = df['Branch Name(purchased from)'].notnull()
    encashed_null = df['Branch Name(encashed from)'].isnull()
    purchased_to_encashed = purchased_not_null & encashed_null
    df.loc[purchased_to_encashed, 'Branch Name(encashed from)'] = df.loc[purchased_to_encashed, 'Branch Name(purchased from)']
```

```python
# Assuming df is your DataFrame
# Make a copy of the original DataFrame
df_modified = df.copy()

# Apply the function to the copied DataFrame
handle_null_values(df_modified)

# Now you can check the changes in the df_modified DataFrame
# For example, you can print the first few rows of the modified DataFrame
df_modified.head()
df_modified.isnull().sum()
```

*Trying to find Outliers in data*

```python
#Z-Score technique
df["Denominations(Encashed)outlier"] = (df["Denominations(Encashed)"] - df["Denominations(Encashed)"].mean()) / df["Denominations(Encashed)"].std()
```

```python
threshold = 3 
outliers = df[df["Denominations(Encashed)outlier"].abs() > threshold]
outliers
```

```python
#Quartile technique
Q1 = df["Denominations(Encashed)outlier"].quantile(0.25)
Q3 = df["Denominations(Encashed)outlier"].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df["Denominations(Encashed)outlier"] < Q1 - 1.5 * IQR) | (df["Denominations(Encashed)outlier"] > Q3 + 1.5 * IQR)]
outliers
```

```python
def create_mapping(original_df, encoded_df, column_name):
    # Get unique values and their frequencies from the original DataFrame
    original_counts = original_df[column_name].value_counts()

    # Get unique values and their corresponding encoded values from the encoded DataFrame
    encoded_counts = encoded_df[column_name].value_counts()

    # Initialize an empty dictionary to store the mappings
    mapping = {}

    # Initialize a counter to keep track of the current encoded value
    current_encoded_value = 1

    # Iterate through the unique values of the column in the original DataFrame
    for value, frequency in original_counts.items():
        # Find the corresponding encoded value with the same frequency
        encoded_value = None
        for idx, enc_value in enumerate(encoded_counts.index):
            if encoded_counts[enc_value] == frequency:
                encoded_value = enc_value
                del encoded_counts[enc_value]  # Remove the selected encoded value from the counts
                break

        if encoded_value is None:
            # If no corresponding encoded value found, use the next available encoded value
            encoded_value = current_encoded_value
            current_encoded_value += 1

        # Map the original value to its encoded value
        mapping[value] = encoded_value  

    return mapping

# Example usage:
person_purchasing_mapping = create_mapping(df, df_new, "Person Purchasing")
```
